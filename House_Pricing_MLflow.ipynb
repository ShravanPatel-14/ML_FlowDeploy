```python
def remove_outliers(df,columns):
    for col in columns:
        q1=df[col].quantile(0.25) #first quartile 25th percentile
        q3=df[col].quantile(0.75) #third quartile 75th percentile
        iqr=q3-q1
        lower_bound=q1-1.5*iqr
        upper_bound=q3*1.5*iqr
        df=df[(df[col]>=lower_bound)&(df[col]<=upper_bound)]
    return df
```

```python
import mlflow
import mlflow.sklearn
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

with mlflow.start_run():    
    # Model object initialization
    RandFor = RandomForestRegressor(n_estimators=5000, max_depth=6, random_state=42)
    GBM = GradientBoostingRegressor(n_estimators=3000, max_depth=6, learning_rate=0.01, min_samples_split=5, random_state=42)
    
    rf_n_estimators = 5000

    # Logging parameters
    mlflow.log_param("rf_n_estimators", rf_n_estimators)
    mlflow.log_param("rf_max_depth", 6)
    mlflow.log_param("gbm_n_estimators", 3000)
    mlflow.log_param("gbm_max_depth", 6)
    mlflow.log_param("gbm_learning_rate", 0.01)
    mlflow.log_param("gbm_minimum_samples_split", 5)
    
    # Training models
    rf = RandFor.fit(x_train, y_train)
    gbm = GBM.fit(x_train, y_train)
    
    # Training scores
    training_score_rf = rf.score(x_train, y_train)  # R^2 score
    training_score_gbm = gbm.score(x_train, y_train)  # R^2 score
    
    # Logging training score
    mlflow.log_metric("rf_training_r2", training_score_rf)
    mlflow.log_metric("gbm_training_r2", training_score_gbm)
    # Value predictions
    y_pred_rf = rf.predict(x_test)
    y_pred_gbm = gbm.predict(x_test)
    # Prediction evaluations
    testing_r2_rf = r2_score(y_test, y_pred_rf)
    testing_r2_gbm = r2_score(y_test, y_pred_gbm)
    testing_mse_rf = mean_squared_error(y_test, y_pred_rf)
    testing_mse_gbm = mean_squared_error(y_test, y_pred_gbm)
    
    # Logging evaluation scores
    mlflow.log_metric("rf_testing_r2", testing_r2_rf)
    mlflow.log_metric("gbm_testing_r2", testing_r2_gbm)
    mlflow.log_metric("rf_testing_mse", testing_mse_rf)
    mlflow.log_metric("gbm_testing_mse", testing_mse_gbm)
    
    # Logging models
    mlflow.sklearn.log_model(rf,"random-forest-regressor")
    mlflow.sklearn.log_model(gbm,"gradient-boosting-regressor")
    
    # Setting tags
    mlflow.set_tag("iteration_name","test_rf_gb_regression")
    mlflow.set_tag("feature_set_version", "loan_default")
```
